```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import math
from scipy.integrate import quad
```

```python
X = pd.read_csv("time_with_loved_ones.csv").to_numpy()
y = pd.read_csv("happiness_level.csv").to_numpy().reshape(-1)  
```

```python
# Add a column of ones to X
X = np.column_stack((np.ones(X.shape[0]), X))

# Perform linear regression
m, n = X.shape
w = np.array([10.0 , 10.0])
learning_rate = 0.001
trajectory = [w.copy()]
for i in range(10000):
    gradient = X.T.dot(X.dot(w) - y) / m
    w -= learning_rate * gradient
    trajectory.append(w.copy())

# Convert trajectory to numpy array for easier manipulation
trajectory = np.array(trajectory)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(trajectory[:, 0], trajectory[:, 1], marker='o', markersize=3, linestyle='-')
plt.title('Gradient Descent Trajectory')
plt.xlabel('Weight 1')
plt.ylabel('Weight 2')
plt.grid(True)
plt.show()
```

```python
# Calculate predicted values
predicted_y = X.dot(w)

# Plot the data and the linear regression line
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 1], y, color='blue', label='Actual data')
plt.plot(X[:, 1], predicted_y, color='red', label='Linear regression')
plt.xlabel('Time with loved ones')
plt.ylabel('Happiness level')
plt.title('Relationship between Time with Loved Ones and Happiness Level')
plt.legend()
plt.grid(True)
plt.show()
```
